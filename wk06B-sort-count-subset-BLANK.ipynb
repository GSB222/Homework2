````python
# Athlete's names A-Z

dfw.sort_values(by= "Athlete Name", ascending = False)
````

````python
## sort city and year
# long ago to now

dfw.sort_values(by = ["City", "Year"], ascending = False)
````

````python
dfs["Season"] = "Summer"
dfs.info()
````

````python
dfw["Season"] = "Winter"
dfw.info()
````

````python
dfs
````

````python
dfs.columns == dfw.columns
list(dfs.columns) == list(dfw.columns)
````

````python
df = pd.concat([dfs, dfw], ignore_index = True)
df
````

````python
df.info()
````

````python
df["Medal"].unique()
````

````python
df["Sex"].unique()
````

````python
df["Sex"] = df["Sex"].astype("category")
df.info()
````

````python
df[["Medal", "Season"]] = df[["Medal", "Season"]].astype("category")
df.info()
````

````python
## df that holds only Sochi
df_sochi = df.query('City == "Sochi"')
df_sochi
````

````python
df_abbot = df.query("`Athlete Name` == 'Jeremy Abbott'")
df_abbot
````

````python
df_1984 = df.query("Year == 1984")
df_1984
````

````python
### return all the competitions between 1920 and 1950 (inclusive)

df.query("1920 <= Year <=1950")
````

````python
#between 1972 and as recent as possible 
#has only tennis
#and only women

df.query("(1972 < Year) &\
(Sport == 'Tennis')\
& (Sex == 'F')")
````

````python
df_gold = df.query("Medal == 'Gold'")
df_gold
````

````python
## 10 countries with the most gold 
df_gold.value_counts("Team").to_frame("Number Gold Medals").head(10)
````

````python
## return as percent

(df_gold.value_counts("Team", normalize = True)*100).to_frame("PCT Gold medals").head(10)
````
